```python
# Import the necessary libraries
import pandas as pd  # pandas is used for working with data tables
import glob         # glob is used to find files by name patterns

# STEP 1: The pattern "*_data.csv" means "find all files that end with '_data.csv'"
# This will find files like: Ahmedabad_data.csv, Delhi_data.csv, Mumbai_data.csv, etc.
city_files = glob.glob("*_data.csv")

# STEP 2: Create an empty list to store all our city data
# We'll put each city's data in this list before combining them
all_cities_data = []

# STEP 3: Read each city file one by one
for file_name in city_files:
    # Read the current city's CSV file into a DataFrame
    # A DataFrame is like a spreadsheet table in Python
    city_df = pd.read_csv(file_name)

    # Add this city's data to our list
    all_cities_data.append(city_df)

    # Optional: Print which file we just read
    print(f"Loaded: {file_name}")

# STEP 4: Combine all city data into one big table
# pd.concat() joins all the DataFrames in our list together
# ignore_index=True makes sure the row numbers are continuous (0, 1, 2, 3...)
combined_data = pd.concat(all_cities_data, ignore_index=True)

# STEP 5: Save the combined data to a new CSV file
# index=False means don't save the row numbers as a separate column
combined_data.to_csv("all_cities_combined.csv", index=False)

# STEP 6: Show us what we accomplished
# len(city_files) = count of how many city files we combined
# len(combined_data) = total number of rows in the final combined file
print(f"SUCCESS: Combined {len(city_files)} city files into one file with {len(combined_data)} total rows")
print("The combined file is saved as: all_cities_combined.csv")
```

```python
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(df)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(df)
display(missing_values.style.background_gradient(cmap='viridis'))
```

```python
# Convert Date to datetime if not already
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.sort_values(by='Date', inplace=True)
```

```python
import plotly.express as px

# Visualize PM2.5 trends for a specific city (e.g., Chennai) with a slider
city_data = df[df['City'] == 'Chennai']
fig = px.line(city_data, x='Date', y='PM2.5', title='PM2.5 Levels in Chennai Over Time',
              template='plotly_dark')

# Add rangeslider
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
# Select numeric columns only
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

fig = px.imshow(corr_matrix, text_auto=True, aspect="auto",
                title="Correlation Matrix of Pollutants",
                color_continuous_scale='RdBu_r')
fig.show()
```

```python
# Compare PM2.5 distributions across top 10 cities
top_cities = df['City'].value_counts().head(10).index
filtered_df = df[df['City'].isin(top_cities)]

fig = px.box(filtered_df, x='City', y='PM2.5', color='City',
             title='Distribution of PM2.5 Levels Across Major Cities',
             template='plotly_white')
fig.show()
```

```python
# Aggregating average AQI per city
city_avg = df.groupby('City')['PM2.5'].mean().reset_index()

# You need to add Lat/Lon data for these cities manually or via a library
# Example Dictionary (Add more cities as needed)
coords = {'Delhi': [28.7041, 77.1025], 'Mumbai': [19.0760, 72.8777], 'Bengaluru': [12.9716, 77.5946]}
city_avg['lat'] = city_avg['City'].map(lambda x: coords.get(x, [None, None])[0])
city_avg['lon'] = city_avg['City'].map(lambda x: coords.get(x, [None, None])[1])

fig = px.scatter_mapbox(city_avg.dropna(), lat="lat", lon="lon", size="PM2.5", color="PM2.5",
                        hover_name="City", zoom=3, mapbox_style="carto-positron",
                        title="Average PM2.5 Intensity by Geography", size_max=50)
fig.show()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

# Pivot data for heatmap
pivot_table = df[df['City'] == 'Delhi'].pivot_table(values='AQI', index='Month', columns='Year', aggfunc='mean')

fig = px.imshow(pivot_table, labels=dict(x="Year", y="Month", color="Average AQI"),
                title="Seasonal AQI Heatmap (Delhi)")
fig.show()
```

```python
# Count occurrences of AQI Buckets per City
sunburst_data = df.groupby(['AQI_Bucket', 'City']).size().reset_index(name='Count')

# Filter for cleaner viz (optional)
sunburst_data = sunburst_data[sunburst_data['Count'] > 50]

fig = px.sunburst(sunburst_data, path=['AQI_Bucket', 'City'], values='Count',
                  title='Composition of AQI Categories by City')
fig.show()
```

```python
# Resample to monthly average for smoother animation
df_monthly = df.groupby(['City', pd.Grouper(key='Date', freq='M')]).mean(numeric_only=True).reset_index()
df_monthly['Date_Str'] = df_monthly['Date'].dt.strftime('%Y-%m')

fig = px.scatter(df_monthly, x="PM2.5", y="NO2", animation_frame="Date_Str",
                 animation_group="City", size="AQI", color="City",
                 hover_name="City", log_x=True, size_max=55,
                 range_x=[10, 1000], range_y=[0, 200],
                 title="Evolution of PM2.5 vs NO2 Over Time (Animated)")
fig.show()
```

```python
fig = px.scatter_3d(df.sample(2000), x='Benzene', y='Toluene', z='Xylene', # Sampling for performance
                    color='AQI_Bucket', opacity=0.7,
                    title='3D Interaction: Benzene, Toluene, and Xylene')
fig.show()
```

```python
import plotly.graph_objects as go

years = df['Date'].dt.year.unique()
fig = go.Figure()

for year in years:
    fig.add_trace(go.Violin(x=df[df['Date'].dt.year == year]['AQI'],
                            name=str(year), side='positive', line_color='black'))

fig.update_traces(orientation='h', width=3, points=False)
fig.update_layout(title='Shift in AQI Distribution by Year', xaxis_title='AQI')
fig.show()
```

```python
from plotly.subplots import make_subplots

sub_fig = make_subplots(specs=[[{"secondary_y": True}]])

city_df = df[df['City'] == 'Ahmedabad'] # Change city as needed

# Trace 1
sub_fig.add_trace(go.Scatter(x=city_df['Date'], y=city_df['PM2.5'], name="PM2.5"), secondary_y=False)

# Trace 2
sub_fig.add_trace(go.Scatter(x=city_df['Date'], y=city_df['AQI'], name="AQI", line=dict(dash='dot')), secondary_y=True)

sub_fig.update_layout(title_text="Relationship between PM2.5 and AQI (Ahmedabad)")
sub_fig.update_yaxes(title_text="PM2.5 Concentration", secondary_y=False)
sub_fig.update_yaxes(title_text="AQI Index", secondary_y=True)
sub_fig.show()
```

```python
import matplotlib.pyplot as plt

Cities = df['City'].unique()
pollutants = ['PM2.5', 'PM10', 'CO', 'O3', 'NO2']

for City in Cities:
    City_data = df[df['City'] == City]
    plt.figure(figsize=(10, 6))
    for pollutant in pollutants:
        plt.bar(pollutant, City_data[pollutant].mean(), label=pollutant)
    plt.title(f'Average Pollutant Levels at Station {City}')
    plt.xlabel('Pollutant')
    plt.ylabel('Average Concentration')
    plt.legend()
    plt.show()
```
