**Palindrome Checker**

**Task: Write a function is_palindrome(text) that returns True if the string is the same when reversed, otherwise False.**

```python
# Solution:
def is_palindrome(text):
  return text == text[::-1]
print(is_palindrome("madam"))
print(is_palindrome("Sebastine"))


```

**Function with Dictionary Lookup**

**Task: Write a function day_of_week(num) that returns the day name for numbers 1‚Äì7 (1 = Monday, 7 = Sunday). Use a dictionary instead of if.**

```python
# Solution:
def day_of_week(num):
  days = {
      1: "Monday",
      2: "Tuesday",
      3: "Wednesday",
      4: "Thursday",
      5: "Friday",
      6: "Saturday",
      7: "Sunday"
  }
  return days.get(num, "Invalid input")
print(day_of_week(1))
print(day_of_week(8))

```

**Multiple Return Values**

**Task: Create a function circle_measurements(radius) that returns both area and circumference of a circle.**

**Use formulas:**

**Area = œÄ √ó r¬≤**

**Circumference = 2 √ó œÄ √ó r**

```python
# Solution:
import math
def circle_measurements(radius):
  area = math.pi * radius ** 2
  circumference = 2 * math.pi * radius
  return area, circumference

a, c = circle_measurements(2)
print("Area:", a)
print("Circumference:", c)


```

**Function with Mathematical Computation**

**Task: Write a function quadratic_roots(a, b, c) that returns the two roots of a quadratic equation ax¬≤ + bx + c = 0. Use the quadratic formula.**

```python
# Solution:
def quadratic_roots(a, b, c):
    d = b**2 - 4*a*c
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    return r1, r2

print(quadratic_roots(3,-5,1))
print(quadratic_roots(1,6,2))


```

**Challenge: Calculator Function**

**Task: Create a function calculator(a, b, operation) where operation can be "add", "subtract", "multiply", or "divide". The function should return the correct result.**

```python
# Solution:
def calculator(a,b,operation):
  operator = {
      "add" : lambda a,b:a+b,
      "sub" : lambda a,b:a-b,
      "div" : lambda a,b:a/b,
      "mul" : lambda a,b:a*b
  }
  return operator.get(operation, lambda a,b:"Invalid Input")(a,b)

print(calculator(2,6,"add"))
print(calculator(39,24,"sub"))
print(calculator(13,73,"mul"))
print(calculator(12,2,"div"))
print(calculator(82,9,"sebastine"))



```
